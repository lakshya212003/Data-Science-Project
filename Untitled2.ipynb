```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.impute import SimpleImputer

%matplotlib inline

# Load the dataset
weather_df = pd.read_csv('kanpur.csv', parse_dates=['date_time'], index_col='date_time')

# Select numerical columns
weather_df_num = weather_df[['maxtempC','mintempC','cloudcover','humidity','tempC', 'sunHour','HeatIndexC', 'precipMM', 'pressure','windspeedKmph']]

# Preprocess data: handle missing values
imputer = SimpleImputer(strategy='mean')
weather_df_num_imputed = pd.DataFrame(imputer.fit_transform(weather_df_num), columns=weather_df_num.columns, index=weather_df_num.index)

# Separate features and target variable
weather_y = weather_df_num_imputed.pop("tempC")
weather_x = weather_df_num_imputed

# Data Visualization
weather_df_num.plot(subplots=True, figsize=(25, 20))
plt.show()

weather_df_num.hist(bins=10, figsize=(15, 15))
plt.show()

weth = weather_df_num['2019':'2020']
plt.scatter(weth.mintempC, weth.tempC)
plt.xlabel("Minimum Temperature")
plt.ylabel("Temperature")
plt.show()

plt.scatter(weth.HeatIndexC, weth.tempC)
plt.xlabel("Heat Index")
plt.ylabel("Temperature")
plt.show()

plt.scatter(weth.pressure, weth.tempC)
plt.xlabel("Pressure")
plt.ylabel("Temperature")
plt.show()

# Split data into training and testing sets
train_X, test_X, train_y, test_y = train_test_split(weather_x, weather_y, test_size=0.2, random_state=4)

# Linear Regression
model_lr = LinearRegression()
model_lr.fit(train_X, train_y)
prediction_lr = model_lr.predict(test_X)

# Data Visualization for Linear Regression
plt.scatter(test_y, prediction_lr)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("Linear Regression")
plt.show()

# Decision Tree Regressor
model_dt = DecisionTreeRegressor(random_state=0)
model_dt.fit(train_X, train_y)
prediction_dt = model_dt.predict(test_X)

# Data Visualization for Decision Tree Regressor
plt.scatter(test_y, prediction_dt)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("Decision Tree Regressor")
plt.show()

# Random Forest Regressor
model_rf = RandomForestRegressor(max_depth=90, random_state=0, n_estimators=100)
model_rf.fit(train_X, train_y)
prediction_rf = model_rf.predict(test_X)

# Data Visualization for Random Forest Regressor
plt.scatter(test_y, prediction_rf)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("Random Forest Regressor")
plt.show()

# K-Nearest Neighbors
model_knn = KNeighborsRegressor()
model_knn.fit(train_X, train_y)
prediction_knn = model_knn.predict(test_X)

# Data Visualization for K-Nearest Neighbors
plt.scatter(test_y, prediction_knn)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("K-Nearest Neighbors")
plt.show()

# Logistic Regression (for the sake of comparison, not typically used for regression tasks)
model_logistic = LogisticRegression()
model_logistic.fit(train_X, train_y)
prediction_logistic = model_logistic.predict(test_X)

# Data Visualization for Logistic Regression
plt.scatter(test_y, prediction_logistic)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.title("Logistic Regression")
plt.show()

# Evaluation
def evaluate(predictions, test_y, model_name):
    mae = mean_absolute_error(test_y, predictions)
    mse = mean_squared_error(test_y, predictions)
    r2 = r2_score(test_y, predictions)
    accuracy = np.sum(np.abs(predictions - test_y) < 2) / len(test_y)  # Define a threshold for accuracy
    print(f"Metrics for {model_name}:")
    print("Mean Absolute Error:", mae)
    print("Mean Squared Error:", mse)
    print("R2-score:", r2)
    print("Accuracy:", accuracy)
    print()

evaluate(prediction_lr, test_y, "Linear Regression")
evaluate(prediction_dt, test_y, "Decision Tree Regressor")
evaluate(prediction_rf, test_y, "Random Forest Regressor")
evaluate(prediction_knn, test_y, "K-Nearest Neighbors")
evaluate(prediction_logistic, test_y, "Logistic Regression")
```
